```python
import gaussian_process, covariance_functions, acquisition_functions, bayesian_optimization, sys, datasets, utils
import numpy as np
from numpy import linspace, meshgrid
import matplotlib.pyplot as p
import scipy.interpolate
from infpy.gp import GaussianProcess
from infpy.gp import SquaredExponentialKernel as SE
kernel = SE([1])
%matplotlib inline

def grid(xy, z, resX=100, resY=100, method='linear'):
    "Convert 3 column data to matplotlib grid"
    x = xy[:,0]
    y = xy[:,1]
    xi = linspace(min(x), max(x), resX)
    yi = linspace(min(y), max(y), resY)
    X, Y = meshgrid(xi, yi)
    #Z = griddata(x, y, z, xi, yi)
    Z = scipy.interpolate.griddata(xy, z, (X, Y), method=method)
    return X, Y, Z

#parameters gaussian process
dimensions = 2
noise = 1
cov_func = covariance_functions.gaussian_kernel
cov_grad = covariance_functions.gaussian_kernel_gradient
width = 30
amp = 10

#initialize gaussian process
gp = gaussian_process.GP_Regressor(dimensions, noise, cov_func, cov_grad, [width, amp])

data = datasets.load_data("population")
p.imshow(np.log(data + 1))
p.show()
data = np.log(data[97:125, 150:200])

xy = np.zeros((data.shape[0] * data.shape[1],2))
z = []
for i in range(data.shape[0]):
    for j in range(data.shape[1]):
        xy[i*data.shape[1]+j,0] = j
        xy[i*data.shape[1]+j,1] = i
        z.append(data[i,j])

X, Y, Z = utils.grid_with_interpolation(xy, z, method='gaussian', smooth=1)
print data.shape
p.contourf(X, Y[::-1], Z)
p.colorbar()
p.show()
p.imshow(Z.reshape((100,100)))
p.show()
data = Z.reshape((100,100))
#print np.max(data)
#print np.min(data)

print 'Max location', np.unravel_index(np.argmax(data), data.shape)


#parameters bayesian optimization
iterations = 40
goal_func = (lambda x: np.abs(np.array([[data[x[0,0],x[0,1]]]]) + np.random.normal(0,0.5)))
                                 #np.random.normal(0,10000,size=x[0].shape)))#lambda x: datasets.toy_func(x) + np.random.normal(0,0.5)#
#xs, ys = np.meshgrid(range(350),range(200))
#p.imshow(np.log(goal_func(ys,xs)))
#p.colorbar()
#p.show()

acqui_func = acquisition_functions.get_function("UCB", kappa=6)
#maybe have something like this

X_range = utils.create_input_space(2, [[0, 100, 1],[0, 100, 1]])
#width, amp, noise
bounds = [(1, 1000), (0.1, 1000), (0.01, 10000)]

#execute bayesian optimization
gp, gp_list, s_list = bayesian_optimization.bayesian_optimization(gp, iterations, goal_func, acqui_func, X_range, bounds, save_every=3)
```
